```python
import pandas as pd

df = pd.read_csv("supermarket_sales.csv")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.month_name()
df["day"] = df["date"].dt.day_name()
```

```python
product_sales = df.groupby("product_line")["total"].sum().sort_values(ascending=False)
product_sales
```

```python
payment_counts = df["payment"].value_counts()
payment_counts
```

**Finding:**
The most commonly used payment method shows customer preference. This helps the supermarket decide which payment options should be supported and promoted.


```python
payment_sales = df.groupby("payment")["total"].sum().sort_values(ascending=False)
payment_sales
```

**Finding:**
The payment method generating the highest total sales contributes the most revenue. This helps identify which payment type is associated with higher spending.

```python
# total sales by day
daily_sales = df.groupby("day")["total"].sum().sort_values(ascending=False)
daily_sales
```

**Finding:**
The day with the highest total sales is the busiest day for the supermarket. This helps management plan staffing, inventory, and promotions more effectively during peak periods.

```python
import matplotlib.pyplot as plt

# branch sales chart
branch_sales.plot(kind="bar")
plt.title("Total Sales by Branch")
plt.ylabel("Sales")
plt.show()
```

**Finding:**
The chart clearly shows the comparison of total sales between branches. It visually confirms that Branch C performs the best.

```python
product_sales.plot(kind="bar")
plt.title("Total Sales by Product Line")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

**Finding:**
The chart shows that Food and Beverages generate the highest sales, while Health and Beauty generates the lowest. This helps the business understand which product categories perform best.

```python
payment_sales.plot(kind="bar")
plt.title("Total Sales by Payment Method")
plt.ylabel("Sales")
plt.show()
```

**Finding:**
The chart shows that Cash generates the highest sales revenue, even though Ewallet is the most frequently used method. This highlights differences between usage and spending behaviour.
